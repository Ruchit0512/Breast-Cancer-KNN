```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score
```

```python
data=pd.read_csv('E:/Data Science SP/Machine Learning/breast-cancer.csv')
data.head()
```

```python
data['diagnosis'].value_counts()
```

```python
data=data.drop(['id','Unnamed: 32'],axis=1)
data.head()
```

```python
data.columns
```

```python
data["diagnosis"]=data["diagnosis"].map({'B':0,'M':1}).astype(int)
data.head()
```

```python
x=data.drop(['diagnosis'],axis=1)
x.head()
```

```python
y=data['diagnosis']
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
k=np.arange(1,25)
train_accuracy=np.empty(len(k))
test_accuracy=np.empty(len(k))
for i,k in enumerate(k):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    train_accuracy[i]=knn.score(x_train,y_train)
    test_accuracy[i]=knn.score(x_test,y_test)
    print(train_accuracy[i])
    print(test_accuracy[i])
```

```python
k=np.arange(1,25)
k
```

```python
k=np.arange(1,25)
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
plt.plot(k,train_accuracy,color='blue',marker='*')
plt.title('Train Accuracy')
plt.xlabel('Numbers Of Neighbors')
plt.ylabel('Train Accuracy')
plt.subplot(1,2,2)
plt.plot(k,test_accuracy,color='r',marker='*')
plt.title('Test Accuracy')
plt.xlabel('Numbers Of Neighbors')
plt.ylabel('Test Accuracy')
plt.show()
```

```python
idx=np.where(test_accuracy==max(test_accuracy))
x=k[idx]
x
```

```python
model=KNeighborsClassifier(n_neighbors=x[0])
model.fit(x_train,y_train)
```

```python
Y_pred=model.predict(x_test)
Y_pred
```

```python
y_test
```

```python
accuracy_score(y_test,Y_pred)
```

```python
model.score(x_train,y_train)
```

```python
confusion_matrix(y_test,Y_pred)
```

```python
pd.crosstab(y_test,Y_pred)
```

```python
X=data.drop(['diagnosis'],axis=1)
Y=data['diagnosis']
```

```python
m=cross_val_score(model,X,Y,cv=10)
m.mean()
```
